```python
# import modules
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# establish parameters and data
pi = np.pi
x = np.arange(0,2*pi, 0.1)
labels = ['xdata', '$\sin(xdata)$', 'example plot']
```

```python
# define function 
def sin(x_data):
    y_data = np.sin(x_data)
    return y_data
```

```python
# make a plot and save it
fig = plt.figure(figsize = (11,8), dpi = 80)
y = sin(x)
plt.plot(x, y)
plt.xlabel(labels[0])
plt.ylabel(labels[1])
plt.title(labels[2])
plt.savefig('example.png')
```
